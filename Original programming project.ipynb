```python
import matplotlib.dates as mdates
import matplotlib
import datetime as datetime
import time

# Creating ordinals so that can plot linear regression of timeseries data, adapted from: https://stackoverflow.com/questions/29308729/can-i-plot-a-linear-regression-with-datetimes-on-the-x-axis-with-seaborn
Coal['date_ordinal'] = pd.to_datetime(Coal['Date']).apply(lambda date: date.toordinal())
x1 = Coal.iloc[0:,2]
y1 = Coal.iloc[0:,0]

def bestfit(x, y):
    x_avg = np.mean(x)
    y_avg = np.mean(y)

    # Subtract the mean of x from all the x1 values and the mean of y from all the y1 values.
    x_zero = x - x_avg
    y_zero = y - y_avg

    # The best m is found by the following calculation.
    m = np.sum(x_zero * y_zero) / np.sum(x_zero * x_zero)
    # Use m from above to calculate the best c.
    c = y_avg - m * x_avg

    return(m, c)

m1, c1 = bestfit(x1,y1)
plt.plot(x1, y1, 'k.', label='Original data')
plt.plot(x1, m1 * x1 + c1, 'b-', label='Best fit line')

sns.set(style="darkgrid")

#plt.xticks(Coal.index,Coal["Date"].values)
plt.xticks(rotation=90)
plt.subplots_adjust(left=0, bottom=0, right=2, top=2, wspace=0.25, hspace=0.25)
plt.show()
```

### Coefficient of Performance (COP)
The coefficient of performance for different fuel types were researched. Each heating system will have a different coefficient of performance which will effect the amount of useful energy emitted from the heating system. In this instance the useful energy would be used to heat the household. The coefficient of performance will affect on the price of energy consumption, if the heating system's COP is low, then the system needs to use more units of fuel to output the correct amount of useful energy than if the heating system had a higher COP. The aim is to choose a heating system that has a high COP but uses a fuel type that outputs a lot of energy per unit and the fuel type is cheap per unit. For the purposes of this project any heating system that uses wood as a fuel can also use coal. Also any heating system that uses natural gas a fuel can also use LPG.
